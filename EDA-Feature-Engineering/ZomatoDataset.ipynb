Data Collection:

Data Storage:
The collected data has been stored in the Comma Separated Value file Zomato.csv. Each restaurant in the dataset is uniquely identified by its Restaurant Id. Every Restaurant contains the following variables:

‚Ä¢ Restaurant Id: Unique id of every restaurant across various cities of the world
‚Ä¢ Restaurant Name: Name of the restaurant
‚Ä¢ Country Code: Country in which restaurant is located
‚Ä¢ City: City in which restaurant is located
‚Ä¢ Address: Address of the restaurant
‚Ä¢ Locality: Location in the city
‚Ä¢ Locality Verbose: Detailed description of the locality
‚Ä¢ Longitude: Longitude coordinate of the restaurant's location
‚Ä¢ Latitude: Latitude coordinate of the restaurant's location
‚Ä¢ Cuisines: Cuisines offered by the restaurant
‚Ä¢ Average Cost for two: Cost for two people in different currencies üë´
‚Ä¢ Currency: Currency of the country
‚Ä¢ Has Table booking: yes/no
‚Ä¢ Has Online delivery: yes/ no
‚Ä¢ Is delivering: yes/ no
‚Ä¢ Switch to order menu: yes/no
‚Ä¢ Price range: range of price of food
‚Ä¢ Aggregate Rating: Average rating out of 5
‚Ä¢ Rating color: depending upon the average rating color
‚Ä¢ Rating text: text on the basis of rating of rating
‚Ä¢ Votes: Number of ratings casted by people

```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.columns
```

```python
df.columns #both have country code. we can combine them
```

```python
combined_df = pd.merge(df, df_country,on='Country Code', how='left')
```

```python
combined_df.head()
```

```python
combined_df.columns
```

```python
combined_df.Country.value_counts()
```

Obsevation: Major customers are in India and major transaction will happen here only

```python
countryNames = combined_df.Country.value_counts().index
```

```python
print(countryNames)
```

```python
countryValues = combined_df.Country.value_counts().values
```

```python
#Creating pie Chart
#Top 3 countries that uses Zomato
plt.pie(countryValues[:3], labels=countryNames[:3], autopct='%1.2f%%')
```

Zomato has maximum transactions are from India followed by USA and followed by UK

```python
ratings = combined_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0 :'No.of Rating'})
```

```python
ratings
```

```python
#Observation
"""
Rating 4.5 to 4.9 - Excellent
Rating 4.0 to 4.4 - V Good
Rating 3.5 to 3.9 - Good
Rating 2.5 to 3.4 - Average
Rating 1.8 to 2.4 - Poor

"""
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot( x ="Aggregate rating", y ="No.of Rating", data = ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot( x ="Aggregate rating", y ="No.of Rating", hue ='Rating color', data = ratings, palette = ['white','Red', 'Orange', 'Yellow', 'Green', 'Green'] )
```

Observation:
Many people didnt rate
Max values are between 2.5 to 3.4

```python
sns.countplot(x="Rating color", data = ratings, palette = ['white','Red', 'Orange', 'Yellow', 'Green', 'Green'] )
```
